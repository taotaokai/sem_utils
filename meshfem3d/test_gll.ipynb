## derivative of Lagrange polynomials
$$\eqalign{
  & {N_i} = \prod\limits_{j = 0,j \ne i}^N {\left( {\xi _i^{GLL} - \xi _j^{GLL}} \right)}   \cr 
  & {\ell _i}\left( \xi  \right) = N_i^{ - 1}\prod\limits_{j = 0,j \ne i}^N {\left( {\xi  - \xi _j^{GLL}} \right)}   \cr 
  & {{\dot \ell }_i}\left( \xi  \right) = N_i^{ - 1}\sum\limits_{k \ne i} {\prod\limits_{m \ne i,k} {\left( {\xi  - \xi _m^{GLL}} \right)} }  \cr} $$

```python
def dLag_dzgll(ngll):
    """ derivative of Lagrange polynomials evaluated at n-point GLL nodes
    """
    z, w = gll_library.zwgljd(ngll,0,0)
    # dLag_i/dz(zj)
    dLag = np.zeros((ngll, ngll))
    ind = np.arange(ngll)
    for i in range(ngll):
        denumerator = np.prod((z[i] - z[ind != i]))
        for j in range(ngll):
            numerator = 0
            for k in range(ngll):
                if k == i: continue
                mask = (ind != i) & (ind != k)
                numerator += np.prod(z[j] - z[mask])
            dLag[i,j] = numerator/denumerator
    return dLag
```

```python
dLag_dzgll(ngll)
```

## GLL nodes and weights

Legendre polynomials given by Rodrigues' formula:
\begin{equation}
  P_n\,(\xi)
  \,=\,\frac{1}{2^n n!}\,\left(\frac{d}{dx}\right)^n\left({\xi}^2\,-\,1\right)^n
\end{equation}

The *Legendre polynomials* can also be calculated via recursion (Bonnet formula) such as:

\begin{split}
    P_0\,(\xi)
    \,&=\,1\\
    P_1\,(\xi)
    \,&=\,\xi\\
    P_n\,(\xi)
    \,&=\,\frac{(2n\,-\,1)\,\xi\, P_{n\,-\,1}\,(\xi)\,-\,(n\,-\,1)\,P_{n\,-\,2}\,(\xi)}{n}\qquad n \geq 2
\end{split}
